## 0. Filter Shapefile by NYC Boundaries, Convert shapefile to geojson 

In this chapter, we will filter 511 events which are located in NYC Boundary. And then, because Sharedstreets API get a geojson file as an input, we have to convert the filtered 511 shapefile to geojson.

```python
import pandas as pd
import geopandas as gpd
from geopandas.tools import sjoin
```

```python
# make sure that you run the '511_00_data_wrangling' notebook to get the '511_events_cleaned.shp'
# import 511 dataset
gdf_511 = gpd.read_file('../data/cleaned_data/511_events_shapefile/511_events_cleaned.shp')
```

```python
# data can be dowloaded from: https://data.cityofnewyork.us/City-Government/Borough-Boundaries-Water-Areas-Included-/tv64-9x69
# import nyc boundary shapefile
gdf_nyc = gpd.read_file('../data/borough_boundaries_w_water/geo_export_b872bcc2-4115-4a61-9581-5cdb4e9449e6.shp')
```

```python
# filter 511 events with NYC boundary
gdf_511_in_nyc = sjoin(gdf_511, gdf_nyc,  op='within')
```

```python
gdf_511_in_nyc.head(3)
```

```python
# reset index and add event_id column
gdf_511_in_nyc = gdf_511_in_nyc.reset_index()
gdf_511_in_nyc = gdf_511_in_nyc.drop('index',axis=1)
gdf_511_in_nyc = gdf_511_in_nyc.reset_index()
gdf_511_in_nyc = gdf_511_in_nyc.rename(columns={'index':'event_id'})
```

```python
gdf_511_in_nyc.head(3)
```

```python
# drop columns from nyc boundary
gdf_511_in_nyc = gdf_511_in_nyc.drop(['index_right','boro_code','boro_name','shape_area','shape_leng'], axis=1)
```

```python
gdf_511_in_nyc.to_file('../data/cleaned_data/511_events_add_id_filtered/511_events_add_id_filtered.shp')
```

```python
gdf_511_in_nyc.to_file('../data/sharedstreets_results/511/before_applying/511_filtered.geojson', driver='GeoJSON')
```

## 1. Data wrangling with Sharedstreets API

To use Sharedstreet API, we have to find optimal number for 'search-radius' parameter. For that, we will look distribution of length of Sharedstreet geometry (segment)

```python
import geopandas as gpd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from fiona.crs import from_epsg
```

```python
gdf_shst_segment = gpd.read_file('../data/sharedstreets_geometry/segment/shst_segment.shp')
```

```python
# change coordinate system from epsg4326 to epsg2263
# epsg2263 is based on us-ft.
gdf_shst_segment.crs = from_epsg(4326)
gdf_shst_segment = gdf_shst_segment.to_crs(epsg=2263)  
```

```python
gdf_shst_segment['length'] = gdf_shst_segment['geometry'].apply(lambda x:x.length)
```

```python
plt.axvline(x=260, c = 'r')
sns.kdeplot(gdf_shst_segment['length'].loc[gdf_shst_segment['length']<1000])
```

We could find a peak point at around 260ft. (the reason will be decribed with more detail). so we decide use 40m (half of 260ft) as a search radius. <br>

## 2. Processing sharedstreet result

Because 'gdf_511_matched' doesn't contain 'event_id' and the other attributes, we need to process the result to get attributes of matched 511 results. We will drop unmatched 511 events from the entire 511 dataset, and concatenate geometries of matached 511 events to it.

```python
import pandas as pd
import geopandas as gpd
```

```python
# make sure that you run the Sharedstreets API to get the result geojson files
# import matched 511 events
gdf_511 = gpd.read_file('../data/sharedstreets_results/511/before_applying/511_filtered.geojson')
gdf_511_matched_result =  gpd.read_file('../data/sharedstreets_results/511/radius40/511_filtered.matched.geojson')
gdf_511_unmatched_result = gpd.read_file('../data/sharedstreets_results/511/radius40/511_filtered.unmatched.geojson')
```

```python
gdf_511.head(3)
```

```python
gdf_511_matched_result.head(3)
```

```python
gdf_511_unmatched_result.head(3)
```

Because 'gdf_511_matched' doesn't contain 'event_id' and the other attributes, we will drop unmatched 511 events in gdf_511.

```python
# extract list of events unmatched 511 events
list_511_unmatched = gdf_511_unmatched_result['event_id'].tolist()
```

```python
# drop the unmatched 511 events
gdf_511_matched = gdf_511.loc[~gdf_511['event_id'].isin(list_511_unmatched)]
```

```python
gdf_511_matched.head(3)
```

To get Sharedstreet geometry Id and mapped geometry, we have to concatenate gdf_511_matched and gdf_511_matched_result

```python
# drop the geometry column to avoid duplicates
gdf_511_matched = gdf_511_matched.drop('geometry', axis=1)
```

```python
# reset index to concatenate 
gdf_511_matched = gdf_511_matched.reset_index().drop('index', axis=1)
gdf_511_matched_result = gdf_511_matched_result.reset_index().drop('index', axis=1)
```

```python
# concatenate geometry columns
gdf_511_matched= pd.concat([gdf_511_matched, gdf_511_matched_result[['geometryId','geometry']]], axis=1)
```

```python
gdf_511_matched.head(3)
```

```python
gdf_511_matched.to_file('../data/cleaned_data/511_matched_0629.geojson', driver='GeoJSON')
```
